```python
img1
```

```python
img = plt.imread(img1)
plt.imshow(img)
```

```python
img.shape
```

RGB 분포 분석

```python
channels = cv2.split(img)
colors = ('b', 'g', 'r')
for (ch, color) in zip (channels, colors):
    hist = cv2.calcHist([ch], [0], None, [256], [0, 256])
    plt.plot(hist, color = color)
plt.show()
```
