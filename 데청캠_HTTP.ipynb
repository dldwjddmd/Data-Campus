```python
# status에 404 쳤을 때 반응 확인
url = 'http://www.httpbin.org/status/403'
resp = get(url=url)

try :    # 트리거
    resp.raise_for_status()
except HTTPError as e:
    print(e.response.status_code)
    print(e.response.reason)
    print(e.response.headers)
    print(e.request.headers)
```
